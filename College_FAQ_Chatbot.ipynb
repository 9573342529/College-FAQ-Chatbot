```python
import re
```

```python
faq_data = {
    "admission": "Admissions are open from June to August. Apply through the college website.",
    "fees": "The annual tuition fee is ₹50,000.",
    "hostel": "Hostel facilities are available for both boys and girls.",
    "courses": "We offer B.Tech, MBA, MCA, and B.Sc programs.",
    "library": "The library is open from 8 AM to 8 PM.",
    "placement": "The college has a placement cell with top recruiters."
}
```

```python
patterns = {
    r".*\badmission\b.*": faq_data["admission"],
    r".*\bfee\b.*": faq_data["fees"],
    r".*\bfees\b.*": faq_data["fees"],
    r".*\bhostel\b.*": faq_data["hostel"],
    r".*\bcourse\b.*": faq_data["courses"],
    r".*\bcourses\b.*": faq_data["courses"],
    r".*\blibrary\b.*": faq_data["library"],
    r".*\bplacement\b.*": faq_data["placement"]
}
```

```python
def get_response(user_input):

    user_input = user_input.lower()

    for pattern, response in patterns.items():
        if re.match(pattern, user_input):
            return response

    return "Sorry, I don't understand your question. Please ask about admissions, fees, hostel, courses, library, or placement."
```

```python
print("College FAQ Chatbot")
print("Type 'bye' to exit")

while True:

    user_input = input("You: ")

    if user_input.lower() == "bye":
        print("Bot: Thank you for visiting. Goodbye!")
        break

    response = get_response(user_input)
    print("Bot:", response)
```

```python
def get_response(user_input):

    user_input = user_input.lower()

    greetings = ["hi", "hello", "hey"]

    if user_input in greetings:
        return "Hello! Welcome to the College FAQ Assistant."

    for pattern, response in patterns.items():
        if re.match(pattern, user_input):
            return response

    return "Sorry, I don't understand your question."
```
